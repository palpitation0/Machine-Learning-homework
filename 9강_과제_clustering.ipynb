# K-means clustering

```
import numpy as np
from sklearn.decomposition import PCA

transformed = PCA(n_components=2).fit_transform(X)
transformed.shape
```

```
from sklearn.cluster import KMeans
km2 = cluster.KMeans(n_clusters=2).fit(transformed)
km3 = cluster.KMeans(n_clusters=3).fit(transformed)
km4 = cluster.KMeans(n_clusters=4).fit(transformed)
```

```
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.scatter(transformed[:,0], transformed[:,1], c=km2.labels_)
plt.title("K=2, J=%.2f" % km2.inertia_)

plt.subplot(132)
plt.scatter(transformed[:,0], transformed[:,1], c=km3.labels_)
plt.title("K=3, J=%.2f" % km3.inertia_)

plt.subplot(133)
plt.scatter(transformed[:,0], transformed[:,1], c=km4.labels_)
plt.title("K=4, J=%.2f" % km4.inertia_)
```

```
X_new = X.to_numpy()
```

```
from sklearn.cluster import KMeans
km2 = cluster.KMeans(n_clusters=2).fit(X_new)
km3 = cluster.KMeans(n_clusters=3).fit(X_new)
km4 = cluster.KMeans(n_clusters=4).fit(X_new)
```

```
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.scatter(X_new[:,0], X_new[:,1], c=km2.labels_)
plt.title("K=2, J=%.2f" % km2.inertia_)

plt.subplot(132)
plt.scatter(X_new[:,0], X_new[:,1], c=km3.labels_)
plt.title("K=3, J=%.2f" % km3.inertia_)

plt.subplot(133)
plt.scatter(X_new[:,0], X_new[:,1], c=km4.labels_)
plt.title("K=4, J=%.2f" % km4.inertia_)
```

# 계층적 군집분석

```
single2 = cluster.AgglomerativeClustering(n_clusters=2, linkage='single').fit(X_new)
single3 = cluster.AgglomerativeClustering(n_clusters=3, linkage='single').fit(X_new)
single4 = cluster.AgglomerativeClustering(n_clusters=4, linkage='single').fit(X_new)
```

```
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.scatter(X_new[:,0],X_new[:,1], c=single2.labels_)
plt.title("K=2")

plt.subplot(132)
plt.scatter(X_new[:,0],X_new[:,1], c=single3.labels_)
plt.title("K=3")

plt.subplot(133)
plt.scatter(X_new[:,0],X_new[:,1], c=single4.labels_)
plt.title("K=4")
```

```
complete2 = cluster.AgglomerativeClustering(n_clusters=2, linkage='complete').fit(X_new)
complete3 = cluster.AgglomerativeClustering(n_clusters=3, linkage='complete').fit(X_new)
complete4 = cluster.AgglomerativeClustering(n_clusters=4, linkage='complete').fit(X_new)
```

```
plt.figure(figsize=(9,3))

plt.subplot(131)
plt.scatter(X_new[:,0],X_new[:,1], c=complete2.labels_)
plt.title("K=2")

plt.subplot(132)
plt.scatter(X_new[:,0],X_new[:,1], c=complete3.labels_)
plt.title("K=3")

plt.subplot(133)
plt.scatter(X_new[:,0],X_new[:,1], c=complete4.labels_)
plt.title("K=4")
```

```
ward2 = cluster.AgglomerativeClustering(n_clusters=2, linkage='ward').fit(X_new)
ward3 = cluster.AgglomerativeClustering(n_clusters=3, linkage='ward').fit(X_new)
ward4 = cluster.AgglomerativeClustering(n_clusters=4, linkage='ward').fit(X_new)
```
